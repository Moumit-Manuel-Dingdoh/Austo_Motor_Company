```python
# outlier checking

num_cols = ['Age','No_of_Dependents', 'Salary', 'Partner_salary', 'Total_salary','Price']

Q1 = df_austo_motor[num_cols].quantile(0.25)
Q3 = df_austo_motor[num_cols].quantile(0.75)

IQR = Q3-Q1

low_whisker = Q1 - 1.5*IQR
up_whisker = Q3 + 1.5*IQR

((df_austo_motor[num_cols] < low_whisker) | (df_austo_motor[num_cols] > up_whisker)).sum()/df_austo_motor.shape[0]*100
```

```python
# Outliers treatment using Clipping/Capping approach

def outliers_treatment(df, col):
    Q1 = df[col].quantile()
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3-Q1

    low_whisker = Q1 - 1.5*IQR
    up_whisker = Q3 + 1.5*IQR
    df[col] = np.clip(df[col], low_whisker, up_whisker)
    return df
```

```python
df_austo_motor = outliers_treatment(df_austo_motor,'No_of_Dependents')
df_austo_motor = outliers_treatment(df_austo_motor,'Total_salary')

fig, ax = plt.subplots(1,2, figsize = (12,6))
sns.boxplot(data=df_austo_motor,x='No_of_Dependents', ax=ax[0])
sns.boxplot(data=df_austo_motor,x='Total_salary', ax=ax[1])

ax[0].set_title('Boxplot for No_of_Dependents')
ax[1].set_title('Boxplot for Total_salary')
plt.tight_layout()
plt.show()
```

### Observations and insights

1. The Age variable is right-skewed without any outliers
2. The No_of_Dependents is normal distribution with outliers present on lower whiskers
3. The distribution for Partner_salary, Salary, Price and Total_salary with a right-skewed without having outliers but, Total_salary is having outliers at the upper whiskers.
4. In catagorical variables, Male has more in count than female and the Personal_loan (Yes/No) are almost equally distributed.
5. The Salaried professions who has completed Post-Graduations, got Married and their Partners are working, and having House_loan were significantly higher.
6. The automobile company manufacture more number of Sedan in compare with SUV and Hatchback.
7. The outliers treatment was done using Clipping/Capping approach to analyse the whole dataset in the next steps to ensure more accurate data analysis.
